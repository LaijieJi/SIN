```python
import string
import matplotlib.pyplot as plt
import igraph as ig
n=5
g = ig.Graph(directed=True, n=n, edges=[[0,1],[0,2],[1,2],[1,3],[2,4],[3,4]])
h = [2,2,1,1,0]
weights = [1,4,1,3,1,1]
g.vs["label"]=[string.ascii_uppercase[i]+"  "+str(h[i]) for i in range(0,n)]
g.vs["label_dist"]=[1e-10 for i in range(0,n)]
g.vs["label_angle"]=[0 for i in range(0,n)]
g.vs["size"]=[25 for i in range(0,n)]
g.vs["color"] = ["white" for i in range(0,n)]
g.es["weight"] = weights
g.es["label"]=[str(i) for i in weights]
layout = g.layout("tree",root=[0]);
ig.plot(g, layout=layout, margin=30, bbox=(0, 0, 200, 200), target="graph.svg")
```
