```python
import matplotlib.pyplot as plt
import igraph as ig
g = ig.Graph(directed=True, n=10, edges=[[0,1],[0,2],[1,3],[1,4],[2,5],[2,6],[3,7],[3,8],[3,9]])
g.vs["label"]=[i for i in range(0,10)]
fig, ax = plt.subplots()
layout = g.layout("tree",root=[0]); layout.rotate(180,dim1=1); ig.plot(g, layout=layout, target=ax);
fig.show()
```
